### Exercise 7: Short-Circuit Evaluation

```python
# TODO: Understand short-circuit evaluation
print("Short-Circuit Evaluation Demo:")

# AND short-circuit: if first is False, don't evaluate second
def expensive_operation():
    print("  Expensive operation called!")
    return True

print("\nTesting AND short-circuit:")
print("False and expensive_operation():")
result1 = False and expensive_operation()  # expensive_operation won't be called
print(f"Result: {result1}")

print("\nTrue and expensive_operation():")
result2 = True and expensive_operation()   # expensive_operation will be called
print(f"Result: {result2}")

# OR short-circuit: if first is True, don't evaluate second
print("\nTesting OR short-circuit:")
print("True or expensive_operation():")
result3 = True or expensive_operation()    # expensive_operation won't be called
print(f"Result: {result3}")

print("\nFalse or expensive_operation():")
result4 = False or expensive_operation()   # expensive_operation will be called
print(f"Result: {result4}")

# TODO: Practical short-circuit example - safe division
def safe_divide(a, b):
    # Check if b is not zero before dividing
    return b != 0 and a / b  # Short-circuit prevents division by zero

print(f"\nSafe Division Examples:")
print(f"safe_divide(10, 2): {safe_divide(10, 2)}")
print(f"safe_divide(10, 0): {safe_divide(10, 0)}")

# TODO: Default value assignment using OR
user_input = ""  # Empty string (falsy)
username = user_input or "Guest"  # Use "Guest" if input is empty
print(f"\nDefault Value Assignment:")
print(f"User input: '{user_input}'")
print(f"Username: '{username}'")

# Try with actual input
user_input2 = "Alice"
username2 = user_input2 or "Guest"
print(f"User input: '{user_input2}'")
print(f"Username: '{username2}'")
```

## üéØ Operator Precedence and Combinations

### Exercise 11: Understanding Operator Precedence

```python
# TODO: Operator precedence in complex expressions
a, b, c = 10, 5, 2

# Arithmetic and comparison precedence
result1 = a + b > c * 3  # What order do operations happen?
result2 = (a + b) > (c * 3)  # Same as above, but explicit
result3 = a + b > c * 3 and b > c  # Multiple operations

print("Operator Precedence Examples:")
print(f"a = {a}, b = {b}, c = {c}")
print(f"\na + b > c * 3: {result1}")
print(f"Step by step: {a} + {b} = {a + b}, {c} * 3 = {c * 3}")
print(f"Then: {a + b} > {c * 3} = {result1}")
print(f"\nExplicit grouping (a + b) > (c * 3): {result2}")
print(f"Complex: a + b > c * 3 and b > c = {result3}")

# TODO: Logical operator precedence
x, y, z = True, False, True

# and has higher precedence than or
result4 = x or y and z  # Same as: x or (y and z)
result5 = (x or y) and z  # Different grouping

print(f"\nLogical Precedence:")
print(f"x = {x}, y = {y}, z = {z}")
print(f"x or y and z: {result4} (same as x or (y and z))")
print(f"(x or y) and z: {result5}")

# TODO: not has highest precedence
result6 = not x and y  # Same as: (not x) and y
result7 = not (x and y)  # Different meaning

print(f"\n'not' precedence:")
print(f"not x and y: {result6} (same as (not x) and y)")
print(f"not (x and y): {result7}")

# TODO: Comparison operators have same precedence, left to right
num = 15
result8 = 10 < num < 20  # Chained comparison
result9 = 10 < num and num < 20  # Equivalent explicit form

print(f"\nChained comparisons:")
print(f"10 < {num} < 20: {result8}")
print(f"10 < {num} and {num} < 20: {result9}")
```

### Exercise 15: Debugging Operator Expressions

```python
# CHALLENGE: Debug these expressions and explain why they give unexpected results

print("üêõ DEBUGGING CHALLENGE")
print("Find the bugs and explain the correct behavior!")
print("="*50)

# Bug 1: Comparison confusion
age = 25
result1 = 18 <= age <= 65 and True  # This looks correct, but...
result2 = (18 <= age <= 65) and True  # What's the difference?

print(f"Bug 1 Analysis:")
print(f"age = {age}")
print(f"18 <= age <= 65 and True: {result1}")
print(f"(18 <= age <= 65) and True: {result2}")
print(f"Are they the same? {result1 == result2}")

# Bug 2: String comparison surprise
value1 = "10"
value2 = "2"
result3 = value1 > value2  # String comparison, not numeric!
result4 = int(value1) > int(value2)  # Numeric comparison

print(f"\nBug 2 Analysis:")
print(f"Comparing strings '10' > '2': {result3}")
print(f"Comparing numbers 10 > 2: {result4}")
print(f"Why different? Strings compare lexicographically (alphabetically)")

# Bug 3: Boolean arithmetic
is_student = True
is_senior = False
discount = is_student + is_senior * 2  # Booleans as numbers!

print(f"\nBug 3 Analysis:")
print(f"is_student = {is_student} (as number: {int(is_student)})")
print(f"is_senior = {is_senior} (as number: {int(is_senior)})")
print(f"Discount calculation: {is_student} + {is_senior} * 2 = {discount}")
print(f"This works because True=1, False=0 in arithmetic")

# Bug 4: Assignment vs comparison
# Commented out because it would cause an error in the notebook
print(f"\nBug 4 Analysis:")
print("# This would be wrong: if x = 5:  # Assignment, not comparison!")
print("# Correct version: if x == 5:   # Comparison")
x = 5
print(f"x == 5: {x == 5}  # Comparison returns True/False")
print(f"x = 5 assigns 5 to x, doesn't return a boolean")

# Bug 5: Chained comparison misunderstanding  
a, b, c = 5, 10, 15
result5 = a < b < c < 20  # All must be true
result6 = a < b and b < c and c < 20  # Equivalent explicit form

print(f"\nBug 5 Analysis:")
print(f"a={a}, b={b}, c={c}")
print(f"Chained: a < b < c < 20 = {result5}")
print(f"Explicit: a < b and b < c and c < 20 = {result6}")

# What if one link breaks?
d = 25
result7 = a < b < d < c  # This should be False
print(f"Broken chain: a < b < {d} < c = {result7}")
print(f"Because {b} < {d} is True, but {d} < {c} is False")

print(f"\nüéØ Key Debugging Tips:")
print(f"1. Use parentheses to make precedence explicit")
print(f"2. Remember string vs numeric comparisons")
print(f"3. Be aware that booleans can act as numbers (True=1, False=0)")
print(f"4. Use == for comparison, = for assignment")
print(f"5. Test each part of chained comparisons individually")
```
