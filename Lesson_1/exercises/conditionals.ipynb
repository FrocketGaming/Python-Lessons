### Exercise 7: Bank Account System

```python
# TODO: Advanced banking system with nested conditions
account_balance = 1250.75
account_type = "premium"  # basic, premium, vip
transaction_type = "withdrawal"  # deposit, withdrawal, transfer
amount = 800.00
daily_limit = 1000.00
todays_withdrawals = 150.00
is_international = False
has_overdraft = True
overdraft_limit = 500.00

print("üè¶ Bank Transaction Processor")
print(f"Account balance: ${account_balance:,.2f}")
print(f"Account type: {account_type}")
print(f"Transaction: {transaction_type} ${amount:,.2f}")
print(f"Daily limit: ${daily_limit:,.2f}")
print(f"Today's withdrawals: ${todays_withdrawals:,.2f}")
print(f"International: {is_international}")
print(f"Overdraft protection: {has_overdraft}")

print("\nüí≥ Processing transaction...")

if transaction_type == "deposit":
    if amount > 0:
        if amount > 10000:
            print("‚ö†Ô∏è Large deposit - will be reported to authorities")
        new_balance = account_balance + amount
        print(f"‚úÖ Deposit successful! New balance: ${new_balance:,.2f}")
    else:
        print("‚ùå Invalid deposit amount")
        
elif transaction_type == "withdrawal":
    # Check daily limit first
    if todays_withdrawals + amount <= daily_limit:
        # Check account balance
        if account_balance >= amount:
            # Regular withdrawal - sufficient funds
            if is_international:
                if account_type == "basic":
                    print("‚ùå International withdrawals not allowed for basic accounts")
                else:
                    fee = amount * 0.03  # 3% international fee
                    total_cost = amount + fee
                    if account_balance >= total_cost:
                        new_balance = account_balance - total_cost
                        print(f"‚úÖ International withdrawal successful!")
                        print(f"   Amount: ${amount:,.2f}")
                        print(f"   Fee: ${fee:.2f}")
                        print(f"   New balance: ${new_balance:,.2f}")
                    else:
                        print(f"‚ùå Insufficient funds (need ${total_cost:,.2f} including fees)")
            else:
                # Domestic withdrawal
                new_balance = account_balance - amount
                print(f"‚úÖ Withdrawal successful! New balance: ${new_balance:,.2f}")
        else:
            # Insufficient funds - check overdraft
            if has_overdraft:
                overdraft_needed = amount - account_balance
                if overdraft_needed <= overdraft_limit:
                    new_balance = account_balance - amount
                    overdraft_fee = 35.00  # Overdraft fee
                    final_balance = new_balance - overdraft_fee
                    print(f"‚ö†Ô∏è Overdraft withdrawal processed")
                    print(f"   Overdraft amount: ${overdraft_needed:.2f}")
                    print(f"   Overdraft fee: ${overdraft_fee:.2f}")
                    print(f"   New balance: ${final_balance:.2f}")
                else:
                    print(f"‚ùå Exceeds overdraft limit (${overdraft_limit:.2f})")
            else:
                print("‚ùå Insufficient funds and no overdraft protection")
    else:
        remaining_daily_limit = daily_limit - todays_withdrawals
        print(f"‚ùå Exceeds daily withdrawal limit")
        print(f"   Remaining limit: ${remaining_daily_limit:.2f}")
        
elif transaction_type == "transfer":
    # TODO: Add your transfer logic here
    if amount > 0:
        if account_balance >= amount:
            # Check transfer limits based on account type
            if account_type == "basic" and amount > 2000:
                print("‚ùå Transfer limit exceeded for basic account ($2,000)")
            elif account_type == "premium" and amount > 10000:
                print("‚ùå Transfer limit exceeded for premium account ($10,000)")
            else:
                # Process transfer
                transfer_fee = 0
                if account_type == "basic" and amount > 500:
                    transfer_fee = 5.00
                
                total_cost = amount + transfer_fee
                if account_balance >= total_cost:
                    new_balance = account_balance - total_cost
                    print(f"‚úÖ Transfer successful!")
                    print(f"   Amount transferred: ${amount:,.2f}")
                    if transfer_fee > 0:
                        print(f"   Transfer fee: ${transfer_fee:.2f}")
                    print(f"   New balance: ${new_balance:,.2f}")
                else:
                    print(f"‚ùå Insufficient funds (need ${total_cost:.2f} including fees)")
        else:
            print("‚ùå Insufficient funds for transfer")
    else:
        print("‚ùå Invalid transfer amount")
else:
    print("‚ùå Unknown transaction type")
```

### Exercise 11: Advanced Decision Tree - Medical Diagnosis System

```python
# CHALLENGE: Create a medical symptom assessment system
# This is for educational purposes only - not real medical advice!

print("üè• SYMPTOM ASSESSMENT SYSTEM (Educational Only)")
print("This is a programming exercise - consult real doctors for medical advice!")
print("=" * 70)

# Patient information
age = 35
temperature = 101.5  # Fahrenheit
has_fever = temperature >= 100.4
has_headache = True
has_cough = False
has_sore_throat = True
has_body_aches = True
has_nausea = False
has_runny_nose = False
difficulty_breathing = False
chest_pain = False
days_with_symptoms = 2
recent_travel = False
contact_with_sick = True
chronic_conditions = ["diabetes"]  # List of conditions
medications = ["metformin"]

print(f"üë§ Patient Profile:")
print(f"Age: {age}, Temperature: {temperature}¬∞F")
print(f"Symptoms: Fever={has_fever}, Headache={has_headache}, Cough={has_cough}")
print(f"          Sore throat={has_sore_throat}, Body aches={has_body_aches}")
print(f"          Nausea={has_nausea}, Runny nose={has_runny_nose}")
print(f"Serious: Breathing issues={difficulty_breathing}, Chest pain={chest_pain}")
print(f"History: Days with symptoms={days_with_symptoms}, Travel={recent_travel}")
print(f"         Contact with sick={contact_with_sick}")
print(f"Medical: Conditions={chronic_conditions}, Medications={medications}")

print(f"\nüîç ASSESSMENT PROCESS:")

# Emergency symptoms check first
emergency_symptoms = difficulty_breathing or chest_pain or temperature >= 103

if emergency_symptoms:
    urgency = "EMERGENCY"
    recommendation = "Seek immediate emergency care"
    print("üö® EMERGENCY SYMPTOMS DETECTED")
    
else:
    # High-risk patient check
    high_risk = age >= 65 or chronic_conditions or (age >= 50 and medications)
    
    if high_risk:
        print("‚ö†Ô∏è High-risk patient identified")
        
        # High-risk assessment
        if has_fever and (has_cough or difficulty_breathing):
            if contact_with_sick or recent_travel:
                urgency = "HIGH"
                recommendation = "Contact doctor immediately - possible infection"
            else:
                urgency = "MODERATE-HIGH"
                recommendation = "Contact doctor today - respiratory symptoms with fever"
        elif has_fever and days_with_symptoms >= 3:
            urgency = "MODERATE"
            recommendation = "Contact doctor - prolonged fever in high-risk patient"
        else:
            urgency = "LOW-MODERATE"
            recommendation = "Monitor symptoms, contact doctor if worsening"
    
    else:
        # Standard patient assessment
        print("‚úÖ Standard risk patient")
        
        # Count symptoms for severity
        symptom_count = sum([
            has_fever, has_headache, has_cough, has_sore_throat,
            has_body_aches, has_nausea, has_runny_nose
        ])
        
        if symptom_count >= 5:
            severity = "High"
        elif symptom_count >= 3:
            severity = "Moderate"
        else:
            severity = "Low"
        
        print(f"Symptom count: {symptom_count}/7, Severity: {severity}")
        
        # Flu-like symptoms pattern
        flu_pattern = has_fever and has_body_aches and (has_headache or has_cough)
        
        # Cold-like symptoms pattern
        cold_pattern = has_runny_nose or (has_sore_throat and not has_fever)
        
        # Stomach bug pattern
        stomach_pattern = has_nausea and not (has_cough or has_runny_nose)
        
        if flu_pattern:
            if contact_with_sick and days_with_symptoms <= 2:
                likely_condition = "Possible flu"
                if severity == "High":
                    urgency = "MODERATE"
                    recommendation = "Contact doctor - severe flu-like symptoms"
                else:
                    urgency = "LOW"
                    recommendation = "Rest, fluids, monitor symptoms. Contact doctor if worsening"
            else:
                likely_condition = "Viral illness"
                urgency = "LOW"
                recommendation = "Rest and fluids, contact doctor if no improvement in 3-5 days"
        
        elif cold_pattern:
            likely_condition = "Possible cold"
            if days_with_symptoms >= 7:
                urgency = "LOW-MODERATE"
                recommendation = "Contact doctor - prolonged cold symptoms"
            else:
                urgency = "LOW"
                recommendation = "Rest, fluids, over-the-counter remedies"
        
        elif stomach_pattern:
            likely_condition = "Possible stomach bug"
            if days_with_symptoms >= 3:
                urgency = "MODERATE"
                recommendation = "Contact doctor - prolonged GI symptoms"
            else:
                urgency = "LOW"
                recommendation = "Clear fluids, bland diet, rest"
        
        else:
            likely_condition = "Non-specific illness"
            if severity == "High" or days_with_symptoms >= 5:
                urgency = "MODERATE"
                recommendation = "Contact doctor for evaluation"
            else:
                urgency = "LOW"
                recommendation = "Monitor symptoms, contact doctor if worsening"

# Final assessment report
print(f"\nüìã ASSESSMENT RESULTS:")
print(f"üè• Urgency Level: {urgency}")
if 'likely_condition' in locals():
    print(f"üî¨ Likely Condition: {likely_condition}")
print(f"üí° Recommendation: {recommendation}")

# Additional care instructions
print(f"\nüè† GENERAL CARE INSTRUCTIONS:")

if has_fever:
    print("üå°Ô∏è For fever: Rest, fluids, fever reducers as needed")

if has_cough:
    print("üò∑ For cough: Stay hydrated, consider throat lozenges")

if has_sore_throat:
    print("üçØ For sore throat: Warm salt water gargle, throat lozenges")

if has_body_aches:
    print("üíä For body aches: Rest, over-the-counter pain relievers")

if has_nausea:
    print("ü•§ For nausea: Clear fluids, bland foods, rest")

print(f"\n‚ö†Ô∏è WHEN TO SEEK IMMEDIATE CARE:")
print(f"   ‚Ä¢ Difficulty breathing or chest pain")
print(f"   ‚Ä¢ Temperature above 103¬∞F")
print(f"   ‚Ä¢ Severe dehydration")
print(f"   ‚Ä¢ Symptoms rapidly worsening")
print(f"   ‚Ä¢ Any concerning changes")

print(f"\n‚öïÔ∏è DISCLAIMER: This is a programming exercise for educational purposes.")
print(f"   Always consult with qualified healthcare providers for medical advice.")
```

```python
# CHALLENGE: Create a text-based adventure game with complex decision trees

print("üéÆ TEXT ADVENTURE: The Mysterious Castle")
print("Navigate through the castle using conditional logic!")
print("=" * 50)

# Player stats
player_health = 100
player_strength = 15
player_intelligence = 12
player_luck = 8
player_gold = 50

# Inventory
has_sword = False
has_key = False
has_potion = True
has_torch = True
has_map = False

# Game state
current_room = "entrance"
dragon_defeated = False
treasure_found = False
princess_rescued = False

print(f"üë§ Player Stats:")
print(f"Health: {player_health}, Strength: {player_strength}, Intelligence: {player_intelligence}, Luck: {player_luck}")
print(f"Gold: {player_gold}")
print(f"Inventory: Sword={has_sword}, Key={has_key}, Potion={has_potion}, Torch={has_torch}, Map={has_map}")

print(f"\nüè∞ You stand before a mysterious castle...")

# Game logic using nested conditionals
if current_room == "entrance":
    print("\nüö™ ENTRANCE HALL")
    print("You see three doors: North (Armory), East (Library), West (Kitchen)")
    
    # Player choice simulation (in real game, this would be user input)
    player_choice = "north"  # Simulating player choosing north
    print(f"You choose to go {player_choice}")
    
    if player_choice == "north":
        current_room = "armory"
        print("\n‚öîÔ∏è ARMORY")
        print("You find an ancient sword glowing with magic!")
        
        if player_strength >= 12:
            has_sword = True
            player_strength += 5  # Sword bonus
            print("‚úÖ You're strong enough to wield the magic sword!")
            print(f"Your strength increases to {player_strength}")
        else:
            print("‚ùå You're not strong enough to lift the heavy sword")
            print("The sword remains on its pedestal")
    
    elif player_choice == "east":
        current_room = "library"
        print("\nüìö LIBRARY")
        print("Ancient books and scrolls fill the shelves")
        
        if player_intelligence >= 10:
            has_map = True
            player_intelligence += 3
            print("‚úÖ You decipher an ancient map of the castle!")
            print(f"Your intelligence increases to {player_intelligence}")
            print("The map reveals secret passages and the dragon's weakness")
        else:
            print("‚ùå The ancient texts are too complex to understand")
            print("You leave the library empty-handed")
    
    elif player_choice == "west":
        current_room = "kitchen"
        print("\nüç≥ KITCHEN")
        print("You smell something delicious cooking...")
        
        if has_torch:
            print("Your torch illuminates a hidden chest!")
            if player_luck >= 7:
                player_gold += 100
                has_key = True
                print(f"üçÄ Lucky find! You discover 100 gold and a mysterious key!")
                print(f"Your gold: {player_gold}")
            else:
                player_gold += 25
                print(f"You find 25 gold pieces. Your gold: {player_gold}")
        else:
            print("Without light, you can't explore properly")
            player_health -= 10  # Stumble in darkness
            print(f"You stumble and hurt yourself. Health: {player_health}")

# Continue to dragon encounter
print(f"\nüêâ DRAGON'S LAIR")
print("You venture deeper and encounter a fearsome dragon!")
print("The dragon blocks the path to the treasure chamber")

# Dragon battle logic
if has_sword and player_strength >= 18:
    print("\n‚öîÔ∏è EPIC BATTLE")
    if has_map:
        print("üìú Your map reveals the dragon's weakness!")
        print("üí• You strike the dragon's vulnerable spot with your magic sword!")
        dragon_defeated = True
        player_health -= 20  # Battle damage
        print(f"üèÜ Dragon defeated! Health remaining: {player_health}")
    else:
        print("‚öîÔ∏è You fight bravely with your sword, but it's a tough battle")
        if player_luck >= 6:
            dragon_defeated = True
            player_health -= 40
            print(f"üçÄ Lucky victory! Dragon defeated! Health: {player_health}")
        else:
            player_health -= 60
            print(f"üíî You're badly wounded but manage to drive off the dragon")
            print(f"Health: {player_health}")
            if player_health <= 0:
                print("üíÄ You have been defeated! Game Over.")

elif player_intelligence >= 15 and has_map:
    print("\nüß† CLEVER SOLUTION")
    print("Using your intelligence and the ancient map, you discover a riddle!")
    print("You solve the riddle and the dragon, impressed, lets you pass peacefully")
    dragon_defeated = True
    print("‚ú® No battle necessary - wisdom triumphs over strength!")

elif has_potion and player_luck >= 8:
    print("\nüß™ ALCHEMICAL SOLUTION")
    print("Your potion contains a sleeping draught!")
    print("üçÄ With great luck, you manage to put the dragon to sleep")
    dragon_defeated = True
    has_potion = False
    print("üò¥ The dragon sleeps peacefully as you sneak past")

else:
    print("\nüò∞ RETREAT")
    print("Without proper equipment or skills, you must retreat!")
    print("The dragon is too powerful to face directly")
    player_health -= 10  # Retreat damage
    print(f"You escape with minor injuries. Health: {player_health}")

# Treasure chamber
if dragon_defeated:
    print(f"\nüíé TREASURE CHAMBER")
    print("With the dragon defeated, you enter the treasure chamber!")
    
    if has_key:
        print("üóùÔ∏è Your mysterious key opens a special chest!")
        treasure_value = 1000
        print(f"üí∞ You find the legendary treasure worth {treasure_value} gold!")
        player_gold += treasure_value
        treasure_found = True
        
        # Bonus: Princess rescue
        print("üë∏ You also find a princess who was held captive!")
        if player_health >= 50:
            princess_rescued = True
            print("‚úÖ You're healthy enough to escort the princess to safety")
        else:
            print("üòû You're too wounded to safely escort the princess")
            print("You promise to return with help")
    
    else:
        treasure_value = 500
        print(f"üí∞ You find treasure worth {treasure_value} gold!")
        player_gold += treasure_value
        treasure_found = True

# Final scoring
print(f"\nüèÅ ADVENTURE COMPLETE!")
print(f"=" * 30)

final_score = 0

if dragon_defeated:
    final_score += 100
    print(f"üêâ Dragon defeated: +100 points")

if treasure_found:
    final_score += 50
    print(f"üíé Treasure found: +50 points")

if princess_rescued:
    final_score += 75
    print(f"üë∏ Princess rescued: +75 points")

final_score += player_gold // 10  # Bonus for gold
final_score += player_health // 5  # Bonus for remaining health

print(f"üí∞ Final gold: {player_gold}")
print(f"‚ù§Ô∏è Final health: {player_health}")
print(f"üèÜ FINAL SCORE: {final_score} points")

# Determine ending
if final_score >= 200:
    ending = "LEGENDARY HERO"
    print(f"üåü {ending}: You are now a legend throughout the land!")
elif final_score >= 150:
    ending = "GREAT ADVENTURER"
    print(f"‚≠ê {ending}: Your deeds will be remembered for generations!")
elif final_score >= 100:
    ending = "SUCCESSFUL QUEST"
    print(f"‚ú® {ending}: You completed your quest admirably!")
else:
    ending = "HUMBLE BEGINNING"
    print(f"üå± {ending}: Every great hero starts somewhere!")

print(f"\nüéÆ Thanks for playing! Your choices and conditional logic determined your fate!")
```
